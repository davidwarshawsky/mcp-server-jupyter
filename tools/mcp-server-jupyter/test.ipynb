```python
import ast

def check_syntax(code: str):
    try:
        ast.parse(code)
        return {"valid": True, "error": None}
    except SyntaxError as e:
        return {
            "valid": False,
            "error": {
                "message": e.msg,
                "line": e.lineno,
                "offset": e.offset,
                "text": e.text
            }
        }

print(check_syntax("print('hello')"))
print(check_syntax("print('hello'"))
```
