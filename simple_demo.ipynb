# Jupyter Notebook Demonstration

This notebook showcases various Jupyter and Python capabilities including data manipulation, visualization, and custom functions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported successfully!")
```

```python
# Create a sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Score': [85.5, 92.0, 78.5, 88.0, 95.5]
}
df = pd.DataFrame(data)
print("DataFrame created!")
print(df)
```

```python
# Display basic statistics
print("Dataset Statistics:")
print(df.describe())
```

## Data Visualization

Below we create a visualization of the Score distribution across our sample dataset.

```python
# Create a simple bar plot
plt.figure(figsize=(10, 5))
plt.bar(df['Name'], df['Score'], color='steelblue')
plt.xlabel('Name')
plt.ylabel('Score')
plt.title('Score Distribution')
plt.ylim(70, 100)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Define a custom function
def calculate_bonus(score):
    """Calculate bonus percentage based on score."""
    if score >= 90:
        return 20
    elif score >= 80:
        return 10
    else:
        return 5

# Apply the function
df['Bonus_Percent'] = df['Score'].apply(calculate_bonus)
print("Bonus percentages calculated:")
print(df[['Name', 'Score', 'Bonus_Percent']])
```

```python
# Perform data transformation
df['Bonus_Amount'] = (df['Score'] * df['Bonus_Percent']) / 100
df['Performance_Level'] = df['Score'].apply(
    lambda x: 'Excellent' if x >= 90 else ('Good' if x >= 80 else 'Needs Improvement')
)
print("\nEnhanced dataset with performance levels:")
print(df)
```

```python
# Summary calculations
average_score = df['Score'].mean()
max_score = df['Score'].max()
min_score = df['Score'].min()

print(f"\nSummary Statistics:")
print(f"Average Score: {average_score:.2f}")
print(f"Maximum Score: {max_score}")
print(f"Minimum Score: {min_score}")
print(f"\nTotal Bonus Amount: ${df['Bonus_Amount'].sum():.2f}")
```
