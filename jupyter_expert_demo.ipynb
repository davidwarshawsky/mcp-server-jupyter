## Small data example
We will build a list and compute a few stats.

```python
# 3) Create a list
values = [3, 7, 2, 9, 4]
values
```

```python
# 4) Compute basic stats
min_val = min(values)
max_val = max(values)
avg_val = sum(values) / len(values)
min_val, max_val, avg_val
```

```python
# 5) Dictionary example
stats = {"min": min_val, "max": max_val, "avg": avg_val}
stats
```

```python
# 6) String formatting
summary = f"min={stats['min']}, max={stats['max']}, avg={stats['avg']:.2f}"
summary
```

```python
# 7) List comprehension
squares = [v * v for v in values]
squares
```

```python
# 8) Simple loop
running_total = 0
for v in values:
    running_total += v
running_total
```

```python
# 9) Basic function
def normalize(vals):
    total = sum(vals)
    return [v / total for v in vals]
normalize(values)
```

```python
# 10) Final check
len(values), len(squares), len(normalize(values))
```
