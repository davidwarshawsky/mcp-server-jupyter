# DuckDB SQL Magic Demo



This notebook demonstrates the `%%duckdb` magic command for running SQL queries on DataFrames.

```python
import pandas as pd

sales = pd.DataFrame({
    "region": ["North", "South", "East", "West"],
    "revenue": [10000, 15000, 12000, 18000]
})
sales
```

```python
%%duckdb
SELECT * FROM sales
WHERE revenue > 12000
ORDER BY revenue DESC
```
